```python
import seaborn as sns
import matplotlib.pyplot as plt

df_subset = df.loc[df['year']==2000]
sns.scatterplot(x='Life Expectancy',y='Total fertility rate',data =df_subset,alpha=0.6)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

x= df['year'].unique()
for i in range(1960,2016,1):
    df_subset = df.loc[df['year']==i]
    sns.scatterplot(x='Life Expectancy',y='Total fertility rate',size='Population',hue='Population',data =df_subset,alpha=0.6,palette="husl")
    plt.legend(loc='lower left')
    sns.set_theme(style="dark")
    plt.axis([10,90,1,10])
    file_name = 'life_exp'+ str(i)+'.png'
    plt.savefig(file_name,dpi=120)
    plt.close()
    
```

```python
import imageio

images = []

for i in range(1960,2016):
    file_name = 'life_exp{}.png'.format(i)
    images.append(imageio.imread(file_name))

imageio.mimsave('output.gif', images, fps=5)
```
