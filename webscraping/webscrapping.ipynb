```python
import pandas as pd
title_list=[]
for title in lyrics_list[0:-8]:
    
    title_list.append(title[36:])
print(title_list)
df = pd.DataFrame({'title':title_list,'url':lyrics_list[0:-8]})
df
```

```python
urls=df.drop_duplicates(['title'])['url'].values
```

```python
prefix_url = "https://www.lyrics.com"

complete_url_list = []

for url in urls:
    complete_url =prefix_url+url
    complete_url_list.append(complete_url)
```

```python
for index,url in enumerate(complete_url_list):
    lyric= requests.get(url).text
    file= open(f' breach{index}.txt',mode='w',encoding='utf-8')
    file.write(lyric)
    file.close()

    
```

```python
breach_corpus=[]
for index,url in enumerate(complete_url_list):
    file = open(f' breach{index}.txt',mode='r',encoding='utf-8')
    lyric=file.read()
    file.close()
    lyric_soup = BeautifulSoup(lyric,'html.parser')

    try:
        lyric_text=lyric_soup.find_all('pre',id="lyric-body-text")[0].text
        breach_corpus.append(lyric_text)
    except:
          pass
```

```python
#x
lyrics_corpus=breach_corpus+pace_corpus
lyrics_corpus
```

```python
corpus=[]

for item in lyrics_corpus:
    cor = item.replace('\n','')
    corpus.append(cor)
corpus
```

```python
#y:
labels = ['Breach'] * len(breach_corpus) + ['Pace'] * len(pace_corpus)
```

```python
labels
```

# Vectorize word counts

```python
vectorizer = CountVectorizer(stop_words='english',max_df=0.8)
X = vectorizer.fit_transform(lyrics_corpus)
X_df = pd.DataFrame(X.todense(), columns=vectorizer.get_feature_names())
X_df.head(3)
```

```python
vectorizer.vocabulary_
```

# Normalise with TF-IDF

```python
tf = TfidfTransformer() 
X_norm = tf.fit_transform(X_df)
```

```python
X_norm
```

```python
X_norm_df=pd.DataFrame(X_norm.todense(), columns=vectorizer.get_feature_names())
X_norm_df.head(3)
```

```python
y=pd.DataFrame(data=labels,columns=['Artist'])
y
```

```python
y['Artist']=y['Artist'].replace('Breach',1)
y['Artist']=y['Artist'].replace('Pace',0)
y.head()
```

```python
X_train, X_test, y_train, y_test=train_test_split(X_norm_df,y,test_size=0.25)
```

```python
X_train.head(3)
```

```python
y_train.head(3)
```

```python
rf = RandomForestClassifier()
rf.fit(X_train,y_train)
y_pred = rf.predict(X_test)
y_pred
```

```python
rf.score(X_train,y_train), rf.score(X_test,y_test)
```

# Hyperparameter Optimisation

```python
#searching for best max_depth
rf = RandomForestClassifier()
rf_params = {"max_depth": range(50,70),"n_estimators": [10, 100, 1000]}

cv = GridSearchCV(rf, rf_params, cv=5,scoring='accuracy', verbose=True)

cv.fit(X_train,y_train)
cv.best_params_, cv.best_score_ 
```

```python
cv.cv_results_
```

```python
res = pd.DataFrame(cv.cv_results_)
res.head(5)
```

```python
from sklearn.metrics import accuracy_score, precision_score, \
recall_score, plot_confusion_matrix, f1_score
```

```python
pred = cv.predict(X_test)
pred_train = cv.predict(X_train)

f1_score(y_train, pred_train),f1_score(y_test, pred) 
```

```python
accuracy_score(y_train, pred_train),accuracy_score(y_test, pred)
```

```python
def print_evaluations(X_test,y_test, cv):
    """Model evaluation metrics"""
    y_pred=cv.predict(X_test)
    y_pred
    print(f'Accuracy: {round(accuracy_score(y_test, y_pred), 3)}')
    print(f'Precision: {round(precision_score(y_test, y_pred), 3)}')
    print(f'Recall: {round(recall_score(y_test, y_pred), 3)}')
    print(f'F1-score: {round(f1_score(y_test, y_pred), 3)}')
    plot_confusion_matrix(cv, X_test, y_test,cmap=plt.cm.Blues,display_labels=['Breach', 'Pace'])
    
```

```python
print_evaluations(X_test,y_test,cv)
```
