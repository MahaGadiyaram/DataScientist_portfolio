```python
#import libraries
import pandas as pd
import json
import re
import os
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import requests
import warnings
import pprint
warnings.simplefilter('ignore')
```

```python
#read all files
for file in os.listdir("./all_ingredients"):
    print(file)
```

```python
#store all csv into a dataframe
df_list = []
for file in os.listdir("./all_ingredients"):
    if file.endswith(".csv"):
        ingredients_df = pd.read_csv(f"./all_ingredients/{file}")
        df_list.append(ingredients_df)
ingredients_df = pd.concat(df_list, sort=True)
```
